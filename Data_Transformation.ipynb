```python
import pandas as pd
import numpy as np
```

# load first dataset

```python
df1= pd.read_csv('used_cars_data.csv')
df1.head(5)
```

# Drop columns taht are not needed( selecting the columns based on kaggle stats on the display of data)

```python
df1 = df1.drop(columns=['bed', 'bed_height','bed_length','combine_fuel_economy','engine_type','fleet','is_certified','is_cpo','is_oemcpo','main_picture_url','theft_title','trimId','vehicle_damage_category'])
```

```python
print(df1.columns)
```

# load second dataset

```python
df2= pd.read_csv('vehicles.csv')
df2.head(5)
```

```python
print(df2.columns)
```

# Renaming the columns in the second dataset to match the column names of first dataset

```python
df2.rename(columns={'region': 'city', 'manufacturer': 'make_name','model': 'model_name','fuel': 'fuel_type','VIN': 'vin','drive':'wheel_system_display','type':'body_type','paint_color':'exterior_color','lat':'latitude','long':'longitude','posting_date':'listed_date'}, inplace=True)
print(df2.columns)
```

# Dropping the unwanted columns in second dataset

```python
df2 = df2.drop(columns=['id', 'url', 'region_url','condition', 'cylinders', 'odometer','title_status', 'size','county','state'])
len(df2.columns)
print(df2.columns)
```

# Performing data standardization before merging both dataset

```python
df2.loc[:,"wheel_system_display"]
```

```python
df2['wheel_system_display'] = df2['wheel_system_display'].str.upper()
df2.loc[:,"wheel_system_display"]
```

# Merging dataset

```python
list = ["city", "price", "year", "make_name","model_name","fuel_type","transmission","vin","wheel_system_display","body_type","exterior_color","description","latitude","longitude","listed_date"]
df_combine = df1.merge(df2,
                   on = list, 
                   how = 'outer')
df_combine.head(5)
```

# Checking for percent of missing values by columns and dropping irrelevant coulmns 

```python
total_rows = len(df_combine)
missing_values = df_combine.isnull().sum()
percentage_missing = (missing_values / total_rows) * 100

# Now, 'percentage_missing' will contain the percentage of missing values for each column
print(percentage_missing)
```

```python
df_combine = df_combine.drop(columns=['cabin', 'image_url'])
```

# Vin number is primary key and missing values need to be dorpped and incorrect vin numbers need to be dropped aswell

```python
df_combine = df_combine.dropna(axis=0, subset=['vin'])
df_combine.head(5)
```

```python
df_combine = df_combine[df_combine['vin'].str.contains('[A-Za-z]')]

print(df_combine)
```

```python
dups_vin = df_combine.pivot_table(columns=['vin'], aggfunc='size')
dups_vin = dups_vin[dups_vin > 1]

print(dups_vin)
```

```python
df_combine = df_combine.drop_duplicates(subset=['vin'])
df_combine['vin'].is_unique
```

# Creating 3 more columns as primary key to split the dataset into a STAR schema model 

```python
start_row = 1
width = 7
prefix1 = 'LOC'
prefix2 = 'DEA'
prefix3 = 'LIS'

df_combine['Location_ID'] = prefix1 + (df_combine.index + start_row).map(str).str.zfill(width)
df_combine['Dealer_ID'] = prefix2 + (df_combine.index + start_row).map(str).str.zfill(width)
df_combine['listing_ID'] = prefix3 + (df_combine.index + start_row).map(str).str.zfill(width)
```

```python
df_combine.head(5)
```

```python
df_combine.to_csv('used_car_dataset.csv', index=False)
```
